データの読み込みと欠損値の確認と処理

```python
from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv("../../Data/FakeNewsdataset.csv", quotechar='"', sep=",") #結合したデータセットを読み込む

#欠損値があるかを確認
missing = df["text"].isna().sum()
print("Missing text:", missing)

#欠損があれば空文字で埋める
if missing > 0:
    df["text"] = df["text"].fillna("")
```

今回のデータには欠損がなかった。重複を除いてデータを訓練用とテスト用に分割

```python
df_unique = df.drop_duplicates(subset=['text'])
texts = df_unique["text"] #テキスト列を抽出
label = df_unique["label"] #ラベル列を抽出

x_train, x_test, y_train, y_test = train_test_split(texts, label, test_size=0.2, random_state=2)
```

labelのデータをfake,trueから0,1に変換

```python
def map_label(label):
    if label == 'fake':
        return 0
    else:
        return 1

def LabelTranslate(y_label):
    y_label_mapped = y_label.apply(map_label) #ラベルを数値に変換
    return y_label_mapped

y_train = LabelTranslate(y_train)
y_test = LabelTranslate(y_test)
```

分割したデータを他のnotebookでも扱えるように保存

```python
import pickle

with open("../../Data/train_test_split.pkl", "wb") as f: #ファイルとして保存
    pickle.dump((x_train, x_test, y_train, y_test), f)
```
